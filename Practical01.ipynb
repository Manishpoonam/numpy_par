# Advance Opration with Business examples


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021,2022,2023,2024]

sale_data = np.array([
    [1, 150000, 180000, 220000, 250000], #CSB
    [2, 120000, 123000, 670000, 670000], #vedara
    [3, 904000, 789000, 546000, 345000], #dhaba
    [4, 560333, 567444, 356000, 345345], #pillu
    [5, 345034, 345245, 767456, 367546], #chai point
])

print("=== Zomato sale Analysis ===")

print("Sale data Shape: ",sale_data.shape )
print("Sale data Size: ", sale_data.size)
print("Sale data Data Type: ", sale_data.dtype)
print("Sale data Dimension: ", sale_data.ndim)
```

# sample data for 3 Restaurant

```python
print("Sample data of 3 Resturant: " ,sale_data[0:3])
print("\n")
print("Sample data of resturant without s.no: \n" ,sale_data[:,1:])
```

# Total sale per year by resutrant 

```python
print("Total sale per year: ",np.sum(sale_data, axis=0)) # but it will take all data including s.no which we dont required
print("\n")
print("Total sale per years: ",np.sum(sale_data[:, 1:], axis=0))
```

# average Sales per restaurant 

```python
print("Average sale", np.mean(sale_data[:, 1:],axis=1 ))
print("\n")

# cumsum
cumsum = np.cumsum(sale_data[:,1:],axis=1)
print("CumSum:", cumsum)
```

# plot on graph

```python
plt.figure(figsize=(8,4))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumlative sale across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# Operations in Vector

```python
vector1 = np.array([1,2,3,4])
vector2 = np.array([5,6,7,8])

print("Addition: ", vector1+vector2)
print("\n Multiplication: ", vector1*vector2)
print("\n Dot Product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\n Angle",angle)


# vectorised operations

res_type = np.array(['birayni', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper = np.vectorize(str.upper)
print("\n Vectorized Upper: ", vectorized_upper(res_type))
```
